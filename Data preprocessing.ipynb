```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid", {"grid.color": ".6"})
```

```python
sit_1_df = pd.read_csv('processed/biogeme_Situation1.txt', sep='\t', index_col='Id')#.drop('Unnamed: 245', axis=1)
sit_2_df = pd.read_csv('processed/biogeme_Situation2.txt', sep='\t', index_col='Id').drop('Unnamed: 245', axis=1)
sit_3_df = pd.read_csv('processed/biogeme_Situation3.txt', sep='\t', index_col='Id').drop('Unnamed: 245', axis=1)

list(sit_1_df.columns) == list(sit_3_df.columns)
```

```python
sit_1_df['Situation'] = int(1)
sit_2_df['Situation'] = int(2)
sit_3_df['Situation'] = int(3)
```

```python
list(sit_1_df.columns)
```

```python
sit_1_df['Age']
```

```python
dfs = [sit_1_df, sit_2_df, sit_3_df]
new_dfs = []

rename_dict = {}

for i in range(1,41):
    rename_dict[f'Window/wall_{i}'] = f'Windowwall_{i}'
        
for df in dfs:
    df = df[(df.Gender != 3) & (df.Gender != 4)]
    df['FREQ_1'] = np.where(df['Frequency'] == 1, 1, 0)
    df['FREQ_1_3'] = np.where(df['Frequency'] == 2, 1, 0)
    df['FREQ_MORE'] = np.where(df['Frequency'] == 3, 1, 0)
    df['MALE'] = np.where(df['Gender'] == 1, 1, 0)
    df['FEMALE'] = np.where(df['Gender'] == 2, 1, 0)
    df['YOUNG'] = np.where(df['Age'] == 1, 1, np.where(df['Age'] == 2, 1, 0))
    df['MIDDLE_AGED'] = np.where(df['Age'] == 3, 1, np.where(df['Age'] == 4, 1, 0))
    df['SENIOR'] = np.where(df['Age'] == 5, 1, np.where(df['Age'] == 6, 1, np.where(df['Age'] == 7, 1, 0)))
    
    # Rename columns
    df.rename(columns=rename_dict, inplace=True)

    new_dfs.append(df)

#list(new_dfs[0].columns)
# sit_1_df = new_dfs[0]
# sit_2_df = new_dfs[1]
# sit_3_df = new_dfs[2]
```
